```python
import random as r
```

```python
class Student:
    def __init__(self, code, name, born, country, industry, st_class):
        self.code = code
        self.name = name
        self.born = born
        self.country = country
        self.industry = industry 
        self.st_class = st_class
    def get_code(self):
        return self.code 
    def set_code(self, code):
        self.code = code
    def get_name(self):
        return self.name 
    def set_name(self, name):
        self.name = name   
    def get_born(self):
        return self.born
    def set_born(self, born):
        self.born = born 
    def get_country(self):
        return self.country
    def set_country(self, country):
        self.country = country
    def get_industry(self):
        return self.industry
    def set_industry(self, industry):
        self.industry = industry
    def get_st_class(self):
        return self.st_class
    def set_st_class(self, st_class):
        self.st_class = st_class
    
    def to_string(self):
        print(f"{self.code}.{self.name} at {self.country}. {self.name} is born {self.born}. \
              {self.name} is learning in {self.industry} at class {self.st_class}")
```

```python
class Manage:
    def __init__(self):
        self.__dict = {}
    
    def print_all_student(self):
        if(len(self.__dict)<1):
            print("List is empty")
        else:
            for student in self.__dict.values():
                student.to_string()
    def add_student(self, name, born, country, industry, st_class):
        code = f"PS{r.randint(1,100)}"
        st = Student(code, name, born, country, industry, st_class)
        self.__dict[code] = st
        pass
    def update_infor(self, code) -> bool:
        try:
            check = False
            for i in self.__dict.keys():
                if(i == code):
                    check = True 

            if(check):
                select = -1
                while(select != 0):
                    print("************************************")
                    print("*1. Update name                    *")
                    print("*2. Update born                    *")
                    print("*3. Update country                 *")
                    print("*4. Update industry                *")
                    print("*5. Update class                   *")
                    print("*0. Cancel                         *")
                    print("************************************")
                    select = int(input("Please enter your updatation: "))
                    if(select == 1):
                        update_name = input("Please enter your name: ")
                        self.__dict[code].set_name(update_name)
                    elif(select == 2):
                        update_born = input("Please enter your born: ")
                        self.__dict[code].set_born(update_born)
                    elif(select == 3):
                        update_country = input("Please enter your country: ")
                        self.__dict[code].set_country(update_country)
                    elif(select == 4):
                        update_industry = input("Please enter your industry: ")
                        self.__dict[code].set_industry(update_industry)
                    elif(select == 5):
                        update_st_class = input("Please enter your class: ")
                        self.__dict[code].set_st_class(update_st_class)
                return True
            else:
                return False
        except:
            return False
    def del_student(self, code):
        check = False
        for i in self.__dict.keys():
            if(i == code):
                check = True 
        if(check):
            del self.__dict[code]
            print("Delete student is successfully")
        else:
            print("Student don't have in list")
    def find_student(self, code):
        for key, value in self.__dict.items():
            if(key == code):
                value.to_string()
                break
    def sort_student(self):
        sorted(self.__dict)
```

```python
if __name__ == "__main__":
    manage = Manage()
    
    
    choose = -1
    while(choose!=0):
        print("*************************************************")
        print("*1. Print list of student                       *")
        print("*2. Add student                                 *")
        print("*3. Update information of student               *")
        print("*4. Delete student in list                      *")
        print("*5. Find student                                *")
        print("*6. Sort student                                *")
        print("*0. Cancel                                      *")
        print("*************************************************")
        choose = int(input("Please enter your choice: "))
        if(choose == 1):
            manage.print_all_student()
        elif(choose == 2):
            name = input("Please enter name: ")
            born = input("Please enter born: ")
            country = input("Please enter country: ")
            industry = input("Please enter industry: ")
            st_class = input("Please enter st_class: ")
            manage.add_student(name, born, country, industry, st_class)
        elif(choose == 3):
            code = input("Please enter code of student: ")
            check = manage.update_infor(code)
            if(check):
                print("You are updated successfully")
            else:
                print("Something was wrong")
        elif(choose == 4):
            code = input("Please enter code of student: ")
            manage.del_student(code)
        elif(choose == 5):
            code = input("Please enter code of student: ")
            manage.find_student(code)
        elif(choose == 6):
            manage.sort_student()
```
