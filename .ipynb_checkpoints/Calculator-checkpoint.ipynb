```python
from tkinter import *

root = Tk()
root.configure(bg="light gray")
root.title("Calculator")
root.geometry("400x200")
row = 0
expression = ""

# def

def calculateResult():
    try:
        global expression
        total = str(eval(expression))
        equation.set(total)
        expression = ""
    except:
        print("Program is error!!!")

def resetExpression():
    global expression
    expression = ""
    equation.set("")
        
def press(num):
    global expression
    equation.set(str(num))
    expression += str(num)

# create Frame

equation = StringVar()

frame = Frame()
frame.pack(fill=X)

Entry(frame, textvariable=equation, width=300).pack()

frame1 = Frame()
frame1.pack(fill=X)

arr = [1, 2, 3, "+", 4, 5, 6, "-", 7, 8, 9, "*", 0, "/", "."]

while True:
    for i in range(4):
        button = Button(frame1, text=str(arr[i]), font="arial 12 bold", width=7, 
               command= lambda num = arr[i]:press(num)).grid(row=row, column=i, padx=10)
    arr = arr[4::]
    row += 1
    if arr[0]==0:
        Button(frame1, text=str(0), font="arial 12 bold", width=7, 
               command=lambda:press(0)).grid(row=row, column=0, padx=10)
        Button(frame1, text=str("/"), font="arial 12 bold", width=7, 
               command=lambda:press("/")).grid(row=row, column=3, padx=10)
        Button(frame1, text=str("."), font="arial 12 bold", width=7, 
               command=lambda:press(".")).grid(row=row+1, column=0, padx=10)
        break    
Button(frame1, text=str("Clear"), font="arial 12 bold", width=7, 
               command=resetExpression).grid(row=row, column=1, padx=10)
Button(frame1, text=str("="), font="arial 12 bold", width=7, 
               command=calculateResult).grid(row=row, column=2, padx=10)
root.mainloop()
```
